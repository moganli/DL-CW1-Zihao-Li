```python
import sklearn
import numpy as np
import os
import matplotlib.pyplot as plt
import pandas as pd
```

**bullet 1**

```python
df= pd.read_csv (r'heart.csv')
X = df[['age', 'sex', 'cp', 'trtbps', 'chol', 'fbs', 'restecg', 'thalachh','exng', 'oldpeak', 'slp', 'caa', 'thall']].to_numpy()
y = df['output'].to_numpy()
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_Y, test_Y = train_test_split(X, y, test_size = 0.2, random_state = 52)
```

**bullet 2**

```python
from sklearn.tree import DecisionTreeClassifier,plot_tree
```

```python
tree_clf = DecisionTreeClassifier(max_depth=3, random_state=42)
tree_clf.fit(train_X,train_Y)
```

```python
plot_tree(tree_clf);
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(tree_clf,train_X,train_Y,cv=10, scoring="accuracy")
```

```python
from sklearn.model_selection import cross_val_predict
train_y_pred = cross_val_predict(tree_clf,X,y, cv=3)
from sklearn.metrics import confusion_matrix
confusion_matrix(y, train_y_pred)
```

TP:96  
FP:42

```python
from sklearn.metrics import precision_score, recall_score
precision_score(y, train_y_pred)
```

```python
recall_score(y, train_y_pred)
```

```python
from sklearn.metrics import f1_score
f1_score(y, train_y_pred)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y, tree_clf.predict_proba(X)[:,1])
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
fpr, tpr, thresholds = roc_curve(y, tree_clf.predict_proba(X)[:,1])
roc_score=roc_auc_score(y, tree_clf.predict_proba(X)[:,1])
plt.figure()
plt.plot(fpr, tpr, label='MNB Model  (area = %0.2f)' % roc_score)
plt.plot([0, 1], [0, 1],'r--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.show()
```

conclusion：If max_depth is greater than 20, it will overfit. Finding a suitable depth is the key point.

**bullet 3**

```python
print("train score:", tree_clf.score(train_X, train_Y))
print("test score:", tree_clf.score(test_X, test_Y))
```

conclusion：Has a certain accuracy rate, but the effect is not very good.   

    Does the decision tree generalize well to new data?
    not really. it may because my dataset dont have too much value.

**bullet 4**

```python
for i in range(2,20) :
    tree_clf = DecisionTreeClassifier(max_depth=i, random_state=42)
    tree_clf.fit(train_X,train_Y)
    print(i,"max_depth, train score:" , tree_clf.score(train_X, train_Y))
    print(i,"max_depth, test score:" , tree_clf.score(test_X, test_Y))
```

Comment: After the maximum depth reaches 8, the accuracy of the evaluation remains 73.8%

```python
splitter = ['random', 'best']
for i in splitter :
    tree_clf = DecisionTreeClassifier(splitter=i,max_depth=3, random_state=42)
    tree_clf.fit(train_X,train_Y)
    print(i,"splitter, train score:" , tree_clf.score(train_X, train_Y))
    print(i,"splitter, test score:" , tree_clf.score(test_X, test_Y))
```

Comment:"best" splitter is better than random

```python
import heapq
vals = np.linspace(0, 0.2, 100)
score = [] 
for i in vals :
    tree_clf = DecisionTreeClassifier(min_impurity_decrease=i,splitter='best',max_depth=3, random_state=42)
    tree_clf.fit(train_X,train_Y)
    print(i,"min_impurity_decrease, train score:" , tree_clf.score(train_X, train_Y))
    print(i,"min_impurity_decrease, test score:" , tree_clf.score(test_X, test_Y))
    score.append(tree_clf.score(test_X, test_Y))
print(heapq.nlargest(1,score)) 
```

comment: 0.01616161616161616   is best for "min_impurity_decrease"

```python

    tree_clf = DecisionTreeClassifier(min_impurity_decrease=0.01616161616161616 ,splitter="best",max_depth=3, random_state=42)
    tree_clf.fit(train_X,train_Y)
    print("train score:" , tree_clf.score(train_X, train_Y))
    print("BEST, test score:" , tree_clf.score(test_X, test_Y))
```

conclusion:    
After adjusting the parameters, the optimal solution is 81.9%    
When the number of samples is small but the sample features are very large, the decision tree is easy to overfit. Generally speaking, it is easier to build a robust model if the number of samples is larger than the number of features.

**bullet 5**

```python
train_X, test_X, train_Y, test_Y = train_test_split(X, y, test_size = 0.3, random_state = 52)
```

```python
tree_clf = DecisionTreeClassifier(min_impurity_decrease=0.01616161616161616 ,splitter="best",max_depth=3, random_state=42)
tree_clf.fit(train_X,train_Y)
print("train score:" , tree_clf.score(train_X, train_Y))
print("30% test date, test score:" , tree_clf.score(test_X, test_Y))
```

Conclusion:  
Because the data set is small, too little data can not complete a good machine learning task, so it will be a good choice to use more data for training.

bullet 7：

If the sample size is small but the sample features are very large, before fitting the decision tree model, it is recommended to do dimensionality reduction, such as principal component analysis (PCA), feature selection (Losso) or independent component analysis (ICA). The dimension of such features will be greatly reduced. Then it will be better to fit the decision tree model.

# bullet 8 :random forests

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_Y, test_Y = train_test_split(X, y, test_size = 0.2, random_state = 52)
```

```python
forests_clf = RandomForestClassifier(max_depth=3, random_state=42)
forests_clf.fit(train_X,train_Y)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(forests_clf,train_X,train_Y,cv=10, scoring="accuracy")
```

```python
from sklearn.model_selection import cross_val_predict
forests_y_pred = cross_val_predict(forests_clf,X,y, cv=3)
from sklearn.metrics import confusion_matrix
confusion_matrix(y, forests_y_pred)
```

```python
from sklearn.metrics import precision_score, recall_score
precision_score(y, forests_y_pred)
```

```python
recall_score(y, forests_y_pred)
```

```python
from sklearn.metrics import f1_score
f1_score(y, forests_y_pred)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y, forests_clf.predict_proba(X)[:,1])
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
fpr, tpr, thresholds = roc_curve(y, forests_clf.predict_proba(X)[:,1])
roc_score=roc_auc_score(y, forests_clf.predict_proba(X)[:,1])
plt.figure()
plt.plot(fpr, tpr, label='MNB Model  (area = %0.2f)' % roc_score)
plt.plot([0, 1], [0, 1],'r--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.show()
```

```python
print("train score:", forests_clf.score(train_X, train_Y))
print("test score:", forests_clf.score(test_X, test_Y))
```

```python
for i in range(2,20) :
    forests_clf = RandomForestClassifier(max_depth=i, random_state=42)
    forests_clf.fit(train_X,train_Y)
    print(i,"max_depth, train score:" , forests_clf.score(train_X, train_Y))
    print(i,"max_depth, test score:" , forests_clf.score(test_X, test_Y))
```

**max_depth=4 is best choise**

```python
vals = np.linspace(0, 0.2, 100)
score1 = [] 
for i in vals :
    forests_clf = RandomForestClassifier(min_impurity_decrease=i,max_depth=4, random_state=42)
    forests_clf.fit(train_X,train_Y)
    print(i,"min_impurity_decrease, train score:" , forests_clf.score(train_X, train_Y))
    print(i,"min_impurity_decrease, test score:" , forests_clf.score(test_X, test_Y))
    score1.append(forests_clf.score(test_X, test_Y))
print(heapq.nlargest(1,score1)) 
```

min_impurity_decrease=0.052525252525252523  is best choise

```python
forests_clf = RandomForestClassifier(min_impurity_decrease=0.052525252525252523 ,max_depth=4, random_state=42)
forests_clf.fit(train_X,train_Y)
print("train score:" , forests_clf.score(train_X, train_Y))
print("BEST, test score:" , forests_clf.score(test_X, test_Y))
```

conclusion:
After adjusting the parameters, the optimal solution is 88.52%
